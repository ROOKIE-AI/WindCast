# 径向基函数对时间进行编码

```python
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
def rbf_encode_time(time, centers, width):  
    """  
    使用径向基函数对时间进行编码。  

    参数:  
        time (float or ndarray): 时间值，可以是单个值或数组（如一天中的时间）。  
        centers (list): 径向基函数的中心点列表。  
        width (float): 径向基函数的宽度（标准差 σ）。  

    返回:  
        ndarray: 编码后的时间特征矩阵。  
    """  
    time = np.array(time).reshape(-1, 1)  # 确保时间是列向量  
    centers = np.array(centers).reshape(1, -1)  # 确保中心点是行向量  
    # 计算 RBF 值  
    rbf_values = np.exp(-((time - centers) ** 2) / (2 * width ** 2))  
    return rbf_values  

# 示例：一天中的时间  
time_points = np.linspace(0, 24, 100)  # 生成从 0 到 24 的 100 个时间点  
centers = [0, 6, 12, 18]  # 设置中心点（午夜、早晨、中午、晚上）  
width = 3  # 设置宽度  

# 对时间点进行 RBF 编码  
rbf_encoded = rbf_encode_time(time_points, centers, width)  

# 绘制 RBF 编码结果  
plt.figure(figsize=(10, 6))  
for i, center in enumerate(centers):  
    plt.plot(time_points, rbf_encoded[:, i], label=f"Center: {center}")  
plt.title("RBF-Encoded Time Features")  
plt.xlabel("Time (hours)")  
plt.ylabel("RBF Value")  
plt.legend()  
plt.grid()  
plt.show()
```
